```python
import numpy as np
import pandas as pd
```

# Series

```python
#Creating Series
#List to Series
s = pd.Series(['a','b','c','d','e'])
s
```

```python
type(s)
```

```python
s = pd.Series(['a','b','c','d','e'], index=[5,6,7,8,9])
s
```

```python
s = pd.Series(['a','b','c','d','e'], index=[10,20,20,40,50])
s
```

```python
# Indexing
s[0]
```

```python
s[10]
```

```python
s[20]
```

```python
s[10:41:1]
```

```python
s[:3:]
```

```python
## Tuple to Series
t = pd.Series((100,200,300,400))
t
```

```python
t = pd.Series((100,200,300,400), index=[10,20,30,40])
t
```

```python
t[10:31:1]
```

```python
## Set to Series
c = pd.Series({1,2,3,4,4,5,5,6})
c
```

```python
#Dictonary to Series
d = pd.Series({'a':1,'b':2,'c':3,'d':4})
d
```

```python
#Dictonary to Series
d = pd.Series({'a':1,'b':2,'c':3,'d':4}, index=['e','f','a','h'])
d
```
