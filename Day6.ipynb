## Creating 2D-Array

```python
import numpy as np
```

```python
a = np.array([[1,2,3],
              [4,5,6]])
a
```

```python
a = np.array([(1,2,3),
              (4,5,6)])
a
```

```python
b = np.array([[1,3,5,7],
              [2,4,6,8],
              [9,2,7,1]])
b
```

```python
#Array attributes
#shape - returns the shape of the array
b.shape
```

```python
b.size
```

```python
b.ndim
```

```python
a = np.array([[[[[1,2,3]]]]])
a.ndim
```

```python
# 
a = np.array([1,2,3], ndmin=32)
a
```

```python
a = np.array([1,2,3],ndmin=33)
a
```

```python
print(np.MAXDIMS)
```
