## wap to count a number of digits and alphabets present inside in entered string

```python
s = input('Enter a string with digits: ')
a = 0
d = 0
for i in s:
    if i.isdigit():
        d+=1
    elif i.isalpha():
        a+=1
print('Number of digits: ', d)
print('Number of alphabets: ', a)
```

### Jump statements (break, continue)

```python
for i in 'Deepankar':
    if i == 'p':
        break
    print(i)
```

### Continur Statement: the continue keyword return control of the iteration to the beginning of the python for loop or python while loop. All remaining lines int the prevailing iteration of the loop are skipped by the continue keyword, which return execution to the beginning of the next iteration ot the loop.

```python
for i in 'Deepankar':
    if i == 'p':
        continue
    print(i)
```

## write a py function to check weather the entered number is even or odd and take the fun name as isEven

```python
def isEven(n):
    if n%2==0:
        print('Even')
    else:
        print('Odd')
```

```python
isEven(56)
```

```python
isEven(31)
```
