## Vector Maths

```python
import numpy as np
```

```python
a = np.array([[1,3],
              [2,4]])
b = np.array([[5,7],
              [6,9]])

print(a)
print(b)
```

```python
# dot
np.dot(a,b)
```

```python
a@b #dot product
```

```python
# cross
np.cross(a,b)
```

```python
# transpose
np.transpose(a)
```

```python
a.T   # transpose
```

```python
#Indexing and Slicing in 1D array
a = np.array([45,78,25,99,66])
a
```

```python
#indexing
#Syntax : var[index_postion]
a[2]
```

```python
a[-3]
```

```python
# Slicing
a[::2]
```

```python
# Indexing and Slicing in 2D Array
b = np.random.randint(1,20,(5,5))
b
```

```python
#indexing
#Syntax : var[row_index, col_index]
b[2,3]
```

```python
b[3] #bydefault row
```

```python
np.where(b==11)
```

```python
np.where(b==17)
```

```python
## Slicing
## Syntax: var[row_slicing, col_slicing]
b[1:4:1, 1:4:1]
```

```python
## Concatination
a = np.random.randint(1, 20, (2,2))
a
```

```python
b = np.random.randint(1, 20,(2,4))
b
```

```python
c = np.random.randint(1, 20,(4,4))
c
```

```python
# axis = 1 ()
np.concatenate((a,b))
```
