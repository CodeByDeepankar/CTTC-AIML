### WAP to print wether the enterd number is odd or even and as well as positive or negetive

```python
n = int(input('Enter a number: '))
if n == 0:
    print('You have entered zero')
else:
    if n > 0:
            if n%2 == 0:
                print('number is positive and even')
            else:
                print('number is positive and odd')
    else:
        if n%2 == 0:
            print('number is negetive and even')
        else:
            print('number is negetive and odd')
```

### WAP to print the grade by taking mark of a student by using if elif else

```python
mark = int(input('Enter your mark: '))
if mark >= 90:
    print('O')
elif mark >= 80:
    print('E')
elif mark >= 70:
    print('A')
elif mark >= 60:
    print('B')
elif mark >= 50:
    print('C')
elif mark >= 40:
    print('D')
else:
    print('Fail')
```

## Loops

### For loop:
        Syntax:
            for i in collection:
            # body of for

```python
for i in 'Deep':
    print(i)
```

```python
for i in [1,2,3,4,5,6]:
    print(i)
```

```python
for i in range(1,7,2):
    print(i)
```

```python
### WAP to print 1 to 100 by using for loop
```

```python
for i in range(1, 101):
    print(i,end=' ')
```

```python
## wap to pritn all the even numbers between 50 to 100 
```

```python
for i in range(50, 101, 2):
    print(i, end=' ')
```

```python
for i in range(50, 101):
    if i % 2 ==0:
        print(i, end=' ')
```

```python
### wap to print the multiplication table of a given number
```

```python
n1 = int(input('Enter a number: '))
for i in range(1, 11):
    print(n1, '*', i, '=',n1*i)
```

## While loop
<pre>
    <b>Syntax:
        Initialization
            while condition:
                #body of while
                increment/decrement
</pre>

```python
i = 1
while i<=7:
    print(i)
    i+=1
```
