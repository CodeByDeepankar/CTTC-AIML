# Arithmetic Operator +, -, *, /, %, //, **

```python
a = 13
b = 3
```

```python
print('a = ', a)
print('b = ', b)
```

```python
print('a + b =', a + b)
print('a - b =', a - b)
print('a * b =', a * b)
print('a / b =', a / b)
print('a % b =', a % b)
print('a // b =', a // b)
print('a ** b =', a ** b)
```
