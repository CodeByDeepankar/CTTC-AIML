# Conditional Statements
<pre>
    <b>Types:
        . Simple if
        . if-else
        . if-elif-else
        . Nested if
</pre>

## Simple if                       
<pre>
    <b>Syntax:
        if condition:
            statement
        
        ^__^  -> indentation = 1 tab space / 4 white space
</pre>

```python
if 5 > 2:
    print(True)
```

```python
if 5 < 2:
    print(True)
```

## if-else                   
<pre>
    <b>Syntax:
        if condition:
            statement1
        else:
            statement2
</pre>

```python
if 5 < 2:
    print(True)
else:
    print(False)
```
